# EDA Notebook


```python
import pandas as pd
df = pd.read_csv("../data/raw/SMSSpamCollection", sep="\t", names=["label", "text"])
df.head()
```

### Class balance

```python
df["label"].value_counts().plot(kind="bar", title="Class Balance (Ham vs Spam)")
```

### Message length distribution

```python
df["length"] = df["text"].str.len()
df.groupby("label")["length"].hist(alpha=0.6, legend=True)
```

### Top words per class (very quick TF counts)

```python
from collections import Counter
ham_words = " ".join(df[df["label"]=="ham"]["text"]).lower().split()
spam_words = " ".join(df[df["label"]=="spam"]["text"]).lower().split()
print(Counter(spam_words).most_common(20))
```
